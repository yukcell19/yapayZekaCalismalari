```python
import requests
from bs4 import BeautifulSoup

url = 'https://medical-recover-teal.vercel.app/'

response = requests.get(url)

# Eğer bağlantı başarılıysa
if response.status_code == 200:
    # HTML içeriğini ayrıştır
    soup = BeautifulSoup(response.content, 'html.parser')

    # Örnek olarak sayfadaki tüm başlıkları (h1, h2, h3, vb.) çek
    for heading in soup.find_all(['h1', 'h2', 'h3']):
        print(heading.text.strip())
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.mackolik.com/canli-sonuclar'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        text = link.text.strip()
        if href:
            print("Bağlantı metni: ", text, "URL:", href)
    print("Toplam bulunan bilgi sayısı: ", len(links), " adet")
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
image = cv2.imread('OpenCVOrnekleri/siberakademi.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görüntünün Gösterimi")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread('OpenCVOrnekleri/trafik-2.png')
image2 = cv2.imread('OpenCVOrnekleri/trafik-3.png')

hsv_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
hsv_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask1_image1 = cv2.inRange(hsv_image1, lower_red, upper_red)
mask1_image2 = cv2.inRange(hsv_image2, lower_red, upper_red)

lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

mask2_image1 = cv2.inRange(hsv_image1, lower_red2, upper_red2)
mask2_image2 = cv2.inRange(hsv_image2, lower_red2, upper_red2)

red_mask_image1 = mask1_image1 + mask2_image1
red_mask_image2 = mask1_image2 + mask2_image2

red_result_image1 = cv2.bitwise_and(image1, image1, mask=red_mask_image1)
red_result_image2 = cv2.bitwise_and(image2, image2, mask=red_mask_image2)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(red_result_image1, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 1')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(red_result_image2, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 2')

plt.show()
```
